# Data analysis of "74% of Food Trucks Don’t Display Prices"

## Cleaning Data Imported from Google Maps

```python
import pandas as pd
df = pd.read_csv('/Users/kunihiro-iwasaki/Desktop/Lede/Project/project1/foodtruck_data.csv')
df.head()
```

```python
df2 = df.drop(columns=['タイトル', 'タグ', 'コメント'])
df2
```

```python
df2['メモ'] = df2['メモ'].str.replace({'、':',', '。':','})
df2
```

## Extract necessary information, such as whether prices are displayed or hidden, and latitude and longitude

```python
df2[['No','price_display','item1','item2','item3','item4','item5']] = df2['メモ'].str.split(',', expand=True)
df2[['urllat','lon']] = df2['URL'].str.split(',', expand=True)
df2[['url1','url2','url3','url4','url5','lat']] = df2['urllat'].str.split('/', expand=True)
df2
cleaned_df = df2[['price_display','lat','lon','item1','item2','item3','item4','item5']]
cleaned_df = cleaned_df.drop(0)
cleaned_df
```

## Calculate the percentage of food trucks that display or do not display prices

```python
cleaned_df['price_display'].value_counts()
```

```python
cleaned_df['price_display'].value_counts(normalize=True)
```
